```python
import pandas as pd

file_path = "nike_products_with_features.csv"
df = pd.read_csv(file_path)

# Create a binary label column based on 'Badge'
df['Label'] = df['Badge'].apply(lambda x: 1 if 'Best Seller' in str(x) else 0)

# Save the labeled dataset
labelled_file_path = "nike_labelled_data.csv"
df.to_csv(labelled_file_path, index=False)

# Display confirmation
print(f"Labeled data saved as {labelled_file_path}")
df.head()
```
